```r
pre_CEN_sf <- read.csv("Zpal_CEN_SubFamily.csv") 

# Load required libraries
library(dplyr)
library(RColorBrewer)

# Define the Spectral color palette (from SF-A to SF-M)

n_colors <- length(unique(pre_CEN_sf$SF))  # Number of unique SF values
color_palette <- colorRampPalette(brewer.pal(11, "Spectral"))(n_colors)  # Extend Set3 to fit your SF categories
# Create a mapping of subfamilies to colors
sf_colors <- data.frame(
  SF = c("SF-A", "SF-B", "SF-C", "SF-D", "SF-E", "SF-F", "SF-G", 
         "SF-H", "SF-I", "SF-J", "SF-K", "SF-L"),
  color = color_palette
)

# Merge the color information with your existing dataframe (CEN_sf)
CEN_sf <- pre_CEN_sf %>%
  left_join(sf_colors, by = "SF")  # Add the color column based on the SF column

CEN_sf
write.csv(CEN_sf,"Zpal_CEN_sf.csv")
```

```r
library(dplyr)

# 创建新的数据框，用于记录每条染色体的边界
sat_boundaries <- CEN_sf %>%
  group_by(Species, Chr) %>%                      # 按照物种和染色体分组
  summarize(
    sat_start = dplyr::first(start),                     # 获取每组的第一行 start 值
    sat_end = dplyr::last(end),.groups = "drop"                           # 获取每组的最后一行 end 值
  )

# 查看生成的边界数据框
sat_boundaries 
write.csv(sat_boundaries,"Zpal_sat_boundary.csv")
```

```r
# 加载必要的库
library(ggplot2)
library(dplyr)
library(tibble)
library(cowplot)

# 创建 plots 目录（如果还没有创建）
if (!dir.exists("./plots/trackers")) {
  dir.create("./plots/trackers", recursive = TRUE)
}

chromosomes <- sprintf("Chr%02d", 1:15)

for (chromosome in chromosomes) {

  # 过滤染色体的数据
  chr_data <- CEN_sf %>% filter(Chr == chromosome)
  
  # 定义SF对应的颜色映射
  sf_colors <- chr_data %>%
    distinct(SF, color) %>%
    deframe()  # 将SF作为命名，颜色为值

  # 绘制 tracker1 并保留图例
  tracker1 <- ggplot(chr_data) +
    geom_rect(aes(xmin = start, xmax = end, ymin = 0.5, ymax = 1, fill = SF)) +
    scale_fill_manual(values = sf_colors) +
    theme_minimal() +
    theme(legend.position = "right")

  # 提取 tracker1 的图例
  legend_tracker1 <- cowplot::get_legend(tracker1)
  
  # 绘制 tracker2 并保留图例
  tracker2 <- ggplot(chr_data) +
    geom_rect(aes(xmin = start, xmax = end, ymin = 0.2, ymax = 0.3, fill = strand)) +
    scale_fill_manual(values = c("+" = "red", "-" = "blue")) +
    theme_minimal() +
    theme(legend.position = "right")

  # 提取 tracker2 的图例
  legend_tracker2 <- cowplot::get_legend(tracker2)
  
  # 绘制 tracker3 并保留图例
  chr_LTR_data <- filtered_LTR_position %>% filter(Chr == chromosome)
  tracker3 <- ggplot(chr_LTR_data) +
    geom_rect(aes(xmin = TE_start, xmax = TE_end, ymin = 0.5, ymax = 1, fill = classification)) +
    scale_fill_manual(values = c("LTR/Gypsy" = "#f8c8d2", "LTR/Copia" = "#66c2a5", "LTR/Solo" = "#8da0cb", 
                                 "LTR/TRIM" = "#fd0505", "LTR/unknown" = "#ffd92f")) +
    theme_minimal() +
    theme(legend.position = "right")

  # 提取 tracker3 的图例
  legend_tracker3 <- cowplot::get_legend(tracker3)

  # 绘制 tracker4 并保留图例
  tracker4 <- ggplot(chr_div_data) +
    geom_rect(aes(xmin = TE_start, xmax = TE_end, ymin = 0, ymax = 1, fill = identity)) +
    scale_fill_gradient(low = "#77ff00", high = "red") +
    theme_minimal() +
    theme(legend.position = "right")

  # 提取 tracker4 的图例
  legend_tracker4 <- cowplot::get_legend(tracker4)

  # 绘制 tracker5 并保留图例
  chr_gene_data <- Zpal_gene_df %>% filter(Chr == chromosome)
  tracker5 <- ggplot(chr_gene_data) +
    geom_rect(aes(xmin = gff_start, xmax = gff_end, ymin = 0.5, ymax = 1, fill = type)) +
    scale_fill_manual(values = c(gene = "#8f1ad8",mRNA="#36971e",CDS="#9aaa35",exon="#0776ec")) +
    theme_minimal() +
    theme(legend.position = "right")

  # 提取 tracker5 的图例
  legend_tracker5 <- cowplot::get_legend(tracker5)

  # 保存每个 tracker 的图例
  ggsave(paste0("./plots/trackers/CEN_", chromosome, "_tracker1_legend.svg"), plot_grid(legend_tracker1), width = 4, height = 4, units = "in", dpi = 300)
  ggsave(paste0("./plots/trackers/CEN_", chromosome, "_tracker2_legend.svg"), plot_grid(legend_tracker2), width = 4, height = 4, units = "in", dpi = 300)
  ggsave(paste0("./plots/trackers/CEN_", chromosome, "_tracker3_legend.svg"), plot_grid(legend_tracker3), width = 4, height = 4, units = "in", dpi = 300)
  ggsave(paste0("./plots/trackers/CEN_", chromosome, "_tracker4_legend.svg"), plot_grid(legend_tracker4), width = 4, height = 4, units = "in", dpi = 300)
  ggsave(paste0("./plots/trackers/CEN_", chromosome, "_tracker5_legend.svg"), plot_grid(legend_tracker5), width = 4, height = 4, units = "in", dpi = 300)
}
```

```r
library(dplyr)
library(ggplot2)
options(repr.plot.width = 5, repr.plot.height = 5)  # 拉长图形

# 1. 去掉 SF 列的 "SF-" 前缀
df <- CEN_sf %>%
  mutate(SF = gsub("SF-", "", SF))  # 只保留字母部分

# 2. 计算每个 SF 在不同 strand 的数量
sf_strand_distribution <- df %>%
  group_by(SF, strand) %>%
  summarise(count = n(), .groups = 'drop')  # 统计每个 SF 在不同 strand 的数量

# 3. 绘制堆叠柱状图展示 SF 与 strand 的关系
p <- ggplot(sf_strand_distribution, aes(x = SF, y = count, fill = strand)) +
  geom_bar(stat = "identity", position = "stack", color = "black") +  # 堆叠柱状图
  scale_fill_manual(values = c("#87CEFA", "#FF9999")) +  # 使用浅蓝色和浅红色
  labs(x = "SF", y = "Count of CentZp", fill = "Strand") +  # 添加标签
  scale_y_continuous(expand = c(0, 0)) +  # 移除 0 和横轴之间的空隙
  theme_minimal() +
  theme(
    axis.text.x = element_text(size = 14, vjust = 0.5, hjust = 0.5),  # 增大标签字体，调整位置
    axis.text.y = element_text(size = 14),  # 调整 Y 轴标签字体大小
    axis.title = element_text(size = 16, face = "bold"),  # 增大坐标轴标题字体大小
    legend.position = "bottom",  # 将图例放在底部
    legend.text = element_text(size = 12),  # 增大图例字体
    panel.grid = element_blank(),  # 移除背景网格
    panel.border = element_rect(color = "black", fill = NA, size = 1),  # 增加数据框的边框
    plot.margin = margin(1, 1, 1, 1, "cm")  # 增加图的边距
  )
p
ggsave("./plots/strand_bias.svg",p,width = 5,height = 5,units = "in",dpi = 300)
```
