```r
pre_CEN_sf <- read.csv("Zpal_CEN_SubFamily.csv") 

# Load required libraries
library(dplyr)
library(RColorBrewer)

# Define the Spectral color palette (from SF-A to SF-M)

n_colors <- length(unique(pre_CEN_sf$SF))  # Number of unique SF values
color_palette <- colorRampPalette(brewer.pal(11, "Spectral"))(n_colors)  # Extend Set3 to fit your SF categories
# Create a mapping of subfamilies to colors
sf_colors <- data.frame(
  SF = c("SF-A", "SF-B", "SF-C", "SF-D", "SF-E", "SF-F", "SF-G", 
         "SF-H", "SF-I", "SF-J", "SF-K", "SF-L", "SF-M"),
  color = color_palette
)

# Merge the color information with your existing dataframe (CEN_sf)
CEN_sf <- pre_CEN_sf %>%
  left_join(sf_colors, by = "SF")  # Add the color column based on the SF column

CEN_sf
write.csv(CEN_sf,"Zpal_CEN_sf.csv")
```

```r
#### subfamily consensus analysis ####

library(dplyr)
library(Biostrings)

# 首先提取每个子家族的唯一序列
SF_seq_unique <- CEN_sf %>%
  group_by(SF) %>%
  summarise(unique_seqs = list(unique(seq)))   # 提取唯一的序列并存储为列表

# 遍历每个子家族并保存为FASTA文件
for(i in 1:nrow(SF_seq_unique)) {
  
  # 获取当前子家族的名称和序列
  sf_name <- SF_seq_unique$SF[i]
  sequences <- SF_seq_unique$unique_seqs[[i]]
  
  # 创建 Biostrings 对象
  dna_sequences <- DNAStringSet(sequences)
  
  # 设置序列的名称
  names(dna_sequences) <- paste0(sf_name, "_seq", 1:length(sequences))
  
  # 导出为 FASTA 文件
  fasta_filename <- paste0(sf_name, "_consensus_sequences.fasta")
  writeXStringSet(dna_sequences, paste0("./sf_align_consensus/", fasta_filename))
  
  # 打印保存信息
  cat("Saved:", fasta_filename, "\n")
}
```

```r
flanking_size <- 1e6

# 创建一个新的列来存储区域信息，统一区域命名
df <- Zpal_TE_position %>%
  mutate(region = case_when(
    TE_start >= CE_start & TE_end <= CE_end ~ 'CEN',
    (TE_start >= (CE_start - flanking_size) & TE_end <= CE_start) | 
    (TE_start >= CE_end & TE_end <= (CE_end + flanking_size)) ~ 'Peri-CEN',
    TRUE ~ 'Non-(peri)-CEN')) %>%
  mutate(region = factor(region, levels = c('CEN', 'Peri-CEN', 'Non-(peri)-CEN')))  # 固定区域顺序

# 创建全基因组的汇总数据
# 全基因组的处理是将所有行的 `identity` 汇总为一个 "whole_genome" 类别
df_whole_genome <- df %>%
  mutate(region = "Genome-wide")  # 全基因组标记

# 将全基因组数据与原始区域数据结合  相当于乘以双倍
df_with_genome <- bind_rows(df, df_whole_genome)
# 加载 dplyr 和 ggplot2 包
library(dplyr)
library(ggplot2)

# 过滤数据，只保留 classification 中包含 "LTR" 的行
filtered_df <- df_with_genome %>%
  filter(grepl("LTR", classification))

# 把全基因组加入 region 中作为新的类别
filtered_df$region <- factor(filtered_df$region, levels = c('CEN', 'Peri-CEN', 'Non-(peri)-CEN', 'Genome-wide'))

```
